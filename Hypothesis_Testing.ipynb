# **Example 1:**

A supermarket plans to launch a loyality program if it results in an aerage spending per shopper of more than $120 per week.

A random sample of 80 shoppers enrolled in the pilot program spent on average of dollar 130 in a week with a standard deviation of $140.

Should the loyality program be launched?

```python
from scipy import stats as stats
import numpy as np
```

```python
n=80
x_bar=130
std=40
mu=120
```

```python
alpha=0.05
```

```python
t_cal=(x_bar-mu)/(std/np.sqrt(n))
t_critical=stats.t.ppf(1-alpha/2,n-1)
```

```python
if t_cal>t_critical:
  print("Launch the Loyality program")
else:
  print("Do not launch the  Loyality program")
```

# **Example 2: One Sample and One Tail t-test**

An outbreak of Salmonella-related illness was attributed to ice-cream produced at a certain factory.

Scientists measured the level of salmmonella in a randomly sampled batches of ice-cream.The levels are 0.593, 0.142, 0.329, 0.691, 0.231, 0.793, 0.519, 0.392, 0.418.

Is their evidence that the mean level of Salmonella in the ice-cream is greater than 0.3?

```python
from scipy import stats as stats
```

```python
mu=0.3
alpha=0.05
data=[0.593, 0.142, 0.329, 0.691, 0.231, 0.793, 0.519, 0.392, 0.418]
p_val=stats.ttest_1samp(data,mu)[1]/2
if p_val<alpha:
  print("Close the Factory")
else:
  print("No need to close the Factory")
```

# **Example 3: Two Sample and Two Tailed Test**

Numbers of control are 91, 87, 99, 77, 88, 91 and numbers of treatment are 101, 110, 103, 93, 99, 104.

Are they significantly different?

```python
from scipy import stats as stats
```

```python
Control=[91, 87, 99, 77, 88, 91]
Treatment=[101, 110, 103, 93, 99, 104]
alpha=0.05
p_val=stats.ttest_ind(Control,Treatment)[1]
if p_val<alpha:
  print("There is significant differnce between Control and Treatment")
else:
  print("There is no significance between Control and Treatment")
```

# **Example 3: Two Proportion Test With Random Number Generation**

In the given data what it is given is from State 1 247 students have enrolled to UG and PG and out of it 36.8% of students have got the job,State 2 308 students have enrolled to UG and PG and out of it 38.9% of students have got the job.

Are the proportion from each state to get the job are different or not?

```python
from scipy import stats as stats
import numpy as np
```

```python
n1=247
n2=308
p1=0.368
p2=0.389
alpha=0.05
pop1=np.random.binomial(1,p1,n1)
pop2=np.random.binomial(1,p2,n2)
p_val=stats.ttest_ind(pop1,pop2)[1]
if p_val<alpha:
  print("There is a significant difference of proportions between state 1 and state 2")
else:
  print("There is no significant difference of proportions between state 1 and state 2")
```
