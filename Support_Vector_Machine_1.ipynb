```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score,classification_report
```

```python
df=pd.read_csv("Non_linear_SVM_Dataset.csv")
```

```python
df.head()
```

```python
df.info()
```

```python
plt.figure(figsize=(8,6))
plt.scatter(df['X1'],df['X2'],c=df['Y'],cmap="viridis")
plt.xlabel("X1")
plt.ylabel("X2")
plt.title("Scatter plot between X1 and X2")
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D
fig=plt.figure(figsize=(10,8))
ax=fig.add_subplot(111,projection="3d")
ax.scatter(df["X1"],df["X2"],df["Y"],c=df["Y"],cmap="viridis")
ax.set_xlabel("X1")
ax.set_ylabel("X2")
ax.set_zlabel("Y")
ax.set_title("Scatterplot between X1,X2 and Y")
plt.show()
```

```python
x=df[["X1","X2"]]
y=df["Y"]
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
svc=SVC()
```

```python
model_svc=svc.fit(x_train,y_train)
```

```python
y_train_pred=model_svc.predict(x_train)
y_test_pred=model_svc.predict(x_test)
```

```python
print("Training score is:",accuracy_score(y_train,y_train_pred))
print("Test score is :",accuracy_score(y_test,y_test_pred))
```

```python
print("Classification report is:",classification_report(y_test,y_test_pred))
```
