```python
import pandas as pd
import numpy as np
df=pd.read_csv("sales.csv")
```

```python
y=df["high"]
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le=LabelEncoder()
```

```python
df["ShelveLoc"]=le.fit_transform(df["ShelveLoc"])
df["Urban"]=le.fit_transform(df["Urban"])
df["US"]=le.fit_transform(df["US"])
```

```python
df.head()
```

```python
x=df.drop("high",axis=1)
```

```python
x
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
ss=StandardScaler()
```

```python
new_x=ss.fit_transform(x)
```

```python
new_x=pd.DataFrame(new_x,columns=x.columns)
```

```python
new_x
```

```python
new_x.drop("Unnamed: 0",axis=1,inplace=True)
```

```python
new_x
```

```python
X=new_x[["CompPrice","Income","Advertising","Population","Price","ShelveLoc","Age","Education","Urban","US"]]
y=df["high"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=42)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dt=DecisionTreeClassifier(criterion="entropy")
```

```python
dt_model=dt.fit(x_train,y_train)
```

```python
y_train_pred=dt_model.predict(x_train)
y_test_pred=dt_model.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score,log_loss
```

```python
print("training score is:",np.round(accuracy_score(y_train,y_train_pred)))
print("Test score is:",np.round(accuracy_score(y_test,y_test_pred)))
```
