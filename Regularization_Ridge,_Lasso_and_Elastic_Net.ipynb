```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
```

```python
df=pd.read_csv("mtcars.csv")
```

```python
df.head()
```

```python
df.info()
```

```python
x=df.drop("mpg",axis=1)
y=df["mpg"]
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x=sc.fit_transform(x)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=42)
```

```python
model=LinearRegression()
```

```python
model_lr=model.fit(x_train,y_train)
```

```python
y_train_pred=model_lr.predict(x_train)
y_test_pred=model_lr.predict(x_test)
```

```python
print(" Train r2 score:",r2_score(y_train,y_train_pred))
print("Test r2 score:",r2_score(y_test,y_test_pred))
```

```python
model_lr.coef_
```

# Ridge

```python
from sklearn.linear_model import Ridge
```

```python
ridge=Ridge(alpha=20)
```

```python
model_ridge=ridge.fit(x_train,y_train)
```

```python
y_train_pred=model_ridge.predict(x_train)
y_test_pred=model_ridge.predict(x_test)
```

```python
print("Train r2 score:",r2_score(y_train,y_train_pred))
print("Test r2 score:",r2_score(y_test,y_test_pred))
```

```python
model_ridge.coef_
```

# Lasso

```python
from sklearn.linear_model import Lasso
```

```python
model=Lasso(alpha=0.5)
```

```python
model_lasso=model.fit(x_train,y_train)
```

```python
y_train_pred=model_lasso.predict(x_train)
y_test_pred=model_lasso.predict(x_test)
```

```python
print("Train r2 score:",r2_score(y_train,y_train_pred))
print("Test r2 score:",r2_score(y_test,y_test_pred))
```

```python
model_lasso.coef_
```

# Elastic Net

```python
from sklearn.linear_model import ElasticNet
```

```python
elasticnet=ElasticNet(alpha=0.5,l1_ratio=0.5)
```

```python
model_elasticnet=elasticnet.fit(x_train,y_train)
```

```python
y_train_pred=model_elasticnet.predict(x_train)
y_test_pred=model_elasticnet.predict(x_test)
```
