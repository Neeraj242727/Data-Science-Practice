 There are 36 people and they are divided into three groups Atkins,GM,SB that is 12 members in each group and they are given a diet plan.

Atkins=(6,1,3,4,2,3,3,2,7,8,10,4)

GM=(4,4,1,7,8,5,3,7,10,4,5,2)

SB=(6,7,5,6,8,7,8,9,6,5,5,12)

These are weights reduced after following that diet plan.

After following that diet plan is the variance between weights among all the three different or not?

```python
Atkins=[6,1,3,4,2,3,3,2,7,8,10,4]
GM=[4,4,1,7,8,5,3,7,10,4,5,2]
SB=[6,7,5,6,8,7,8,9,6,5,5,12]
alpha=0.05
import pandas as pd
data=pd.DataFrame({'Atkins':Atkins,'GM':GM,'SB':SB})
from scipy import stats as stats
p_val=stats.f_oneway(data['Atkins'],data['GM'],data['SB'])[1]
p_val
if p_val<alpha:
  print("There is some difference in the weight proportions between the groups")
else:
  print("There is no difference in the weight proportions between the groups")
```
